# Task4

```python
#extracting the min and max year from data set

youtube_data['created_year'].min()
```

```python
#max 
youtube_data['created_year'].max()
```

```python
#drpoing the year 1970 from dataset
youtube_data.drop(youtube_data.loc[youtube_data["created_year"]==1970].index,inplace=True)
```

# Task5

```python
# Replacing the null values with a value.

youtube_data['Category'].fillna('Not Specified',inplace=True)
youtube_data['Country'].fillna('Not Specified',inplace=True)
youtube_data['Abbreviation'].fillna('Not Specified',inplace=True)
youtube_data['channel_type'].fillna('Not Specified',inplace=True)
youtube_data['created_month'].fillna('Not Specified',inplace=True)

youtube_data['video_views_rank'].fillna(0,inplace=True)
youtube_data['Country_rank'].fillna(0,inplace=True)
youtube_data['Channel_type_rank'].fillna(0,inplace=True)
youtube_data['video_views_for_the_last_30_days'].fillna(0,inplace=True)
youtube_data['subscribers_for_last_30_days'].fillna(0,inplace=True)


youtube_data['created_year'].fillna(0,inplace=True)
youtube_data['created_date'].fillna(0,inplace=True)
youtube_data['Gross tertiary education enrollment (%)'].fillna(0,inplace=True)
youtube_data['Population'].fillna(0,inplace=True)
youtube_data['Unemployment rate'].fillna(0,inplace=True)
youtube_data['Urban_population'].fillna(0,inplace=True)
youtube_data['Latitude'].fillna(0,inplace=True)
youtube_data['Longitude'].fillna(0,inplace=True)

```

```python
#null count after replacing with a value
youtube_data.isnull().sum()
```

# Task7 #Performing bivariate and multivariate analysis on categorical columns with the help of graphs/charts, as mentioned below.

```python
#Top10 youtube_channel with respect to highest_yearly_earnings

Top_yt_hy_channel=youtube_data.groupby(by=['Youtuber'])['highest_yearly_earnings'].max()
Top_yt_hy_channel

Top10_Yt_hy_channel_earning=Top_yt_hy_channel.sort_values(ascending=False).head(10)
Top10_Yt_hy_channel_earning.reset_index()
```

```python
#plot Bar graph for visualization of Highest_yearly_earnings youtubers 

Top10_Yt_hy_channel_earning.reset_index().plot(kind='barh',x='Youtuber',y='highest_yearly_earnings')
plt.title('Youtuber vs highest_yearly_earnings ')
plt.ylabel("Youtuber")             
plt.xlabel("highest_yearly_earnings")
plt.show()
```

# Task10

```python
#Top10 categories with respect to subscribers and highest_yearly_earnings, views

top_categories=youtube_data.groupby("Category")['highest_yearly_earnings',"subscribers","video views"].max()
Top10_categories=top_categories.sort_values(by=["subscribers"],ascending=False).head(10)
Top10_categories.reset_index()
```

```python
# ploting the line graph Category vs subscribers count

plt.plot(Top10_categories.reset_index()['Category'],Top10_categories.reset_index()['subscribers'], marker='o')
plt.title ('Category vs subscribers')
plt.ylabel ('subscribers count')
plt.xticks(rotation= 90)
plt.show()

# ploting the line graph Category vs highest_yearly_earnings 
plt.plot(Top10_categories.reset_index()['Category'],Top10_categories.reset_index()['highest_yearly_earnings'], marker='o')
plt.title ('Category vs highest_yearly_earnings')
plt.ylabel ('highest_yearly_earnings ')
plt.xticks(rotation= 90)
plt.show()

# ploting the line graph Category vs  video_views count
plt.plot(Top10_categories.reset_index()['Category'],Top10_categories.reset_index()['video views'], marker='o')
plt.title ('Category vs video views')
plt.ylabel ('video_views count')
plt.xticks(rotation= 90)
plt.show()

```
